# Mod3_Final Assessment - Data Visualization

In this phase, the cleaned dataset is analyzed through visualizations in order to answer key business questions in a clear and intuitive way.

Each visualization focuses on a specific aspect of customer behavior or profile, such as monthly flight activity, the relationship between distance and loyalty points, customer distribution across provinces, salary differences by education level, loyalty card proportions, and demographic composition.

Appropriate chart types are selected depending on the nature of the variables involved, ensuring that each visualization effectively communicates the underlying patterns in the data.

The objective of this phase is not only to present the data visually, but also to interpret the results in a concise and business-oriented manner.

```python
# Import libraries

import pandas as pd
from pathlib import Path
import sys

# Add the src folder to the Python path
sys.path.append(str(Path("../src").resolve()))

# Import the visualization class
from support_visualizations import Visualizations

# Display all columns when printing DataFrames
pd.set_option("display.max_columns", None)
```

```python
# Load the cleaned dataset from Phase 1

df = pd.read_csv("../files/processed/merged_dataset_v2.csv")

df.head()
```

```python
# Create visualization object

vis = Visualizations(df)
```

## 1. How are flights booked distributed by month throughout the year?

In this visualization, we analyze how the number of flights booked varies across the months of the year.

A bar chart is selected because it allows for a clear comparison of discrete categories (months) and helps identify patterns such as seasonality or peak travel periods.

```python
vis.flights_booked_by_month()
```

The chart shows that flight bookings are not evenly distributed throughout the year.

Booking activity increases notably in the middle of the year, reaching its peak in **July**, with also high values in **June**, **August**, and **December**. In contrast, **January** and **February** show the lowest booking volumes.

Overall, this pattern suggests seasonality in customer travel behavior, with higher demand during summer months and at the end of the year.

## 2. Is there a relationship between flight distance and accumulated points?

This visualization explores the relationship between the distance traveled and the number of loyalty points accumulated.

A scatter plot is used because it is the most appropriate chart to analyze relationships between two continuous numerical variables and to identify potential correlations or patterns.

```python
vis.distance_vs_points()
```

The scatter plot shows a very strong positive relationship between distance and points accumulated.

As distance increases, the number of accumulated points also rises almost proportionally, which is consistent with the logic of a loyalty program where rewards are linked to travel activity.

This suggests that flight distance is one of the main drivers of points accumulation in the dataset.

## 3. How are customers distributed across provinces?

This visualization shows the number of customers in each province.

A bar chart is used because it allows for an easy comparison between categories and clearly highlights which provinces concentrate the highest number of customers.

```python
vis.customers_by_province()
```

The chart shows that customer distribution is concentrated in a small number of provinces.

**Ontario** has the highest number of customers by a clear margin, followed by **British Columbia** and **Quebec**. The remaining provinces show considerably lower customer counts.

This suggests that the loyalty program has a stronger presence in specific geographical areas rather than being evenly distributed across the country.

## 4. How does average salary compare across education levels?

This visualization compares the average salary across different education levels.

A bar chart is selected because it is effective for comparing aggregated numerical values across categorical groups.

```python
vis.salary_by_education()
```

The chart shows clear differences in average salary across education levels.

Customers with a **Doctor** degree have by far the highest average salary, followed by those with a **Master** degree. In contrast, **Bachelor**, **College**, and **High School or Below** show noticeably lower average salary levels.

Overall, the visualization suggests a positive relationship between educational attainment and income, helping define the socioeconomic profile of the customer base.

## 5. What is the proportion of customers by loyalty card type?

This visualization represents the proportion of customers holding each type of loyalty card.

A pie chart is used because it is suitable for showing relative proportions within a whole and allows for an intuitive understanding of distribution.

```python
vis.loyalty_card_proportion()
```

The pie chart shows that the customer base is concentrated in a limited number of loyalty card types.

**Star** is the most common card type, followed by **Nova**, while **Aurora** represents the smallest share of customers.

This indicates that most customers are concentrated in the upper two tiers of the loyalty program rather than being evenly distributed across all card types.

## 6. How are customers distributed by marital status and gender?

This visualization shows the distribution of customers across marital status categories, segmented by gender.

A grouped bar chart (countplot with hue) is used because it allows for comparison across two categorical variables simultaneously.

```python
vis.marital_status_by_gender()
```

The chart shows how customers are distributed across marital status categories, with a breakdown by gender.

The **Married** group is the most represented, while **Single** and **Divorced** contain fewer customers. Gender distribution appears relatively balanced within all marital status categories, with no major differences between men and women.

This provides a clear view of the demographic composition of the customer base and suggests that marital status has a stronger visible variation than gender in this dataset.

## Conclusion

The visual analysis provides a clearer and more intuitive understanding of the main patterns in the dataset.

The results show that flight activity is seasonal, with booking peaks concentrated in the middle of the year and another high point in December. A very strong positive relationship between distance and accumulated points confirms that loyalty rewards are closely tied to travel behavior.

Geographically, customers are concentrated in a few provinces, especially Ontario. The salary analysis also reveals clear differences across education levels, with higher education generally associated with higher average income.

In addition, the loyalty card distribution shows that most customers are concentrated in the Star and Nova tiers, while the demographic analysis indicates that the Married group is the most represented and that gender distribution remains relatively balanced across categories.

Overall, the visualizations complement the statistical analysis by translating the main data patterns into findings that are easier to interpret and communicate.
