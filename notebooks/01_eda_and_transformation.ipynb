## 2. Data Cleaning

```python
# Create a copy of the merged dataset to perform the cleaning process safely

df_copy = df.copy()

df_copy.head(2)
```

We start the cleaning phase by reviewing the missing values identified during the exploration step.

**Cancellation Year** and **Cancellation Month** contain a high percentage of null values. These missing values do not represent data quality issues, but rather indicate that the customer has not cancelled their membership.

Since these variables are not required for the subsequent analysis and their null values carry meaningful information, they are preserved without modification.

```python
# Review the Salary variable, since it contains missing values and will be used in later analysis

df_copy["Salary"].describe()
```

Salary contains missing values and some invalid negative values, which are not realistic for this variable. Therefore, negative values are treated as data errors and replaced with missing values before imputation.

```python
# Identify rows with negative salary values

df_copy[df_copy["Salary"] < 0] 
```

A total of 480 entries contain negative values in **Salary**. As mentioned, negative salary values are considered data errors, so they are treated as missing values.

Instead of modifying them arbitrarily, they are replaced with **NaN** and later imputed using a more reliable approach.

```python
# Replace invalid negative salary values with NaN

df_copy.loc[df_copy["Salary"] < 0, "Salary"] = np.nan
```

```python
# Impute missing Salary values using the median within each education group. This approach is more realistic because salary is likely to vary by education level, and the median is robust to extreme values

df_copy["Salary"] = df_copy.groupby("Education")["Salary"].transform(lambda x: x.fillna(x.median()))

# Fill any remaining missing values with the overall median

df_copy["Salary"] = df_copy["Salary"].fillna(df_copy["Salary"].median())
```

Salary was first imputed using the median within each education group. However, some groups contained only missing values, preventing proper imputation.

To address this, the remaining missing values were filled using the overall median of the dataset.

The next step is to treat exact duplicate rows. A total of 1,864 duplicated entries were identified in the dataset.

```python
# Display duplicated rows, including both original and repeated entries

duplicates = df_copy[df_copy.duplicated(keep=False)]

duplicates.head()
```

```python
# Remove exact duplicated rows

df_copy = df_copy.drop_duplicates()

# Confirm that no exact duplicated rows remain

print(f"Remaining duplicated rows: {df_copy.duplicated().sum()}")
```

Duplicate rows were identified as exact repetitions across all columns. To better understand the issue, duplicated entries were inspected, confirming that they are identical observations.

Many of these duplicated rows contain zero values across activity-related variables, suggesting that they may originate from data collection or processing issues rather than real customer behavior.

Since these records do not provide additional information and may bias the analysis, they are removed to ensure data consistency and reliability.

After removing exact duplicate rows, repeated records are still found for some combinations of **Loyalty Number**, **Year**, and **Month**.

These records are not identical, as they contain different values in activity-related variables. This indicates that multiple entries exist for the same customer and period.

Since there is no clear and consistent rule to aggregate or reconcile these differences, these records are not modified. Instead, they are documented as a data quality limitation of the dataset.

```python
# Check how many customer-period combinations have multiple records

df_copy.groupby(["Loyalty Number", "Year", "Month"]).size().value_counts()
```

Most customer-period combinations have a single record (399,622 cases). However, 2,060 combinations contain two records, and a small number of cases (6) contain three records.

This confirms that while the majority of the dataset is consistent, there are some instances where multiple records exist for the same customer and period, indicating potential data inconsistencies.

Given that these cases represent a small proportion of the dataset, their impact on the overall analysis is expected to be limited.
