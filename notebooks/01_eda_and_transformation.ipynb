## 2. Data Cleaning

```python
# First, create a copy of our merged dataset for safety

df_copy = df.copy()

df_copy.head(2)
```

We will start deleting or tratando our missing values.
If we check about next steps of this proyect, we don't need to use Cancellation Year or Cancellation Month for any visualization. So, we are going to check quickly if our hipothesis is right and there are nulls because they are still clients. If that, we will replace NaN for 0, to keep the data type.

Cancellation Year and Cancellation Month contain a high percentage of null values. These missing values do not indicate data quality issues, but rather that the customer has not cancelled their membership.

Since no further analysis involving cancellation dates is required, these null values are preserved as they carry meaningful information and do not need to be modified.

```python
# Salary will be needed for further visualizations. We are going to analyze the missing values

df_copy["Salary"].describe()
```

Salary contains missing values and some invalid negative values, which are not realistic for this variable.

```python
# Analyzing negative values

df_copy[df_copy["Salary"] < 0] 
```

480 entries contains negative values on Salary. As mentioned, negative salary values are considered data errors, so they are treated as missing values. Instead of modifying them arbitrarily, they are replaced with NaN and later imputed using a more reliable approach.

```python
df_copy.loc[df_copy["Salary"] < 0, "Salary"] = np.nan
```

```python
# Missing values are imputed using the median salary within each education group
# This approach allows for a more realistic estimation, as salary is likely to vary depending on the level of education, and the median is robust to outliers
# Median is preferred over the mean to reduce the impact of extreme values

df_copy["Salary"] = df_copy.groupby("Education")["Salary"].transform(lambda x: x.fillna(x.median()))
```
